```python
# IQR
# calculate interquartile range
data = data_cleaned
list_numeric_device = list(data[data['is_binary_device'] == False].device_id.unique()).copy()
iqr_column_names = ['device_id', 'q25', 'q75', 'lower', 'upper']
df_iqr = pd.DataFrame(columns = iqr_column_names)
for dev_id in list_numeric_device:
  data_by_device = data[data['device_id'] == dev_id]
  _mean = data_by_device.device_value.mean()
  _max = data_by_device.device_value.max()
  _min = data_by_device.device_value.min()# ( cai nay co the tinh duoc +- alpha*std)
  _std = data_by_device.device_value.std()
  q25, q75 = data_by_device.device_value.quantile(.25), data_by_device.device_value.quantile(.75)
  iqr = q75 - q25
  cut_off = iqr * 1.5
  lower, upper = q25 - cut_off, q75 + cut_off
  uni_cut_off = 3*_std
  uni_lower, uni_upper = _mean - uni_cut_off, _mean + uni_cut_off 
  df_iqr = df_iqr.append({'device_id': dev_id, '_mean': _mean, 'std': _std, 'uni_lower': uni_lower, 'uni_upper': uni_upper,
                          'q25': q25, 'q75': q75, 'lower': lower, 'upper': upper, '_min': _min, '_max': _max}, ignore_index=True)
df_iqr
```

```python

def process_read_data(data):
  if data.empty:
    raise Exception("Data is empty")
  # Extract devices and state
  device_list = data['device_id'].unique()
  device_state = set()
  for device in device_list:
    max_state = data[data['device_id'] == device]['mask'].max()
    if pd.isna(max_state):
      max_state = "ON"
    device_state.add(device + '_' + str(max_state))
  return device_state


def check_anomaly(state_assoc_rule, data):
    try:
        candidate = []
        extracted_state =  process_read_data(data)
        row_labels, column_labels = list(extracted_state.copy()), list(extracted_state.copy())
        table = pd.DataFrame(0, columns=column_labels, index=row_labels)
        # Note: check if there are more than 1 device
        if(len(table.columns.values.tolist()) <= 1):
            return candidate
        for rule in zip(*state_assoc_rule.to_dict('list').values()):
            antecedant = set(rule[0])
            contecedant = set(rule[1])
            if (contecedant.issubset(set(table.columns.values))) and (antecedant.issubset(set(table.columns.values))):
                for con in contecedant:
                    for ant in antecedant:
                        table.loc[ant, con] = table.loc[ant, con] + 1
#         print(f'table: {table}')
        candidate_row = (table != 0).any(axis=1)
        candidate_col = (table != 0).any(axis=0)
        ORoperator = candidate_row | candidate_col
        candidate = ORoperator.index[ORoperator == False].tolist()
        # print(f'candidate row: {candidate_row} \n candidate col: {candidate_col}')
        return candidate
    except Exception as err:
        print(err)

# candidate, extracted_state = check_anomaly(state_assoc_rules, data_test)
```
